**Información de Shannon:** Cantidad de información de un punto con respecto a su probabilidad.



 La entropía de Shannon se basa en la idea de que la cantidad de información en un evento o en una distribución de probabilidad está relacionada con la sorpresa asociada a ese evento.

 La entropía mide cuán inesperado o sorprendente es un resultado.

```python
# librerias
import numpy as np
import matplotlib.pyplot as plt
```

```python
# variables conocidas

# probabilidades de un dado con 6 caras
probabilidades_dado = np.array([1/6, 1/6, 1/6, 1/6, 1/6, 1/6])

# Calcular la entropía para cada probabilidad por separado
entropias_cara = -np.log2(probabilidades_dado) # h(x) = -log2(p(x))
entropia_total = np.sum(entropias_cara)

# entropías individuales
for i, entropia in enumerate(entropias_cara):
    print(f'Entropía (Cara {i + 1}): {entropia:.5f} bits')

# entropia total
print(f'Entropía total del dado de 6 caras: {entropia_total:.5f} bits')

# Graficar la distribución de probabilidad
etiquetas = ['Cara 1', 'Cara 2', 'Cara 3', 'Cara 4', 'Cara 5', 'Cara 6']
plt.bar(etiquetas, probabilidades_dado)
plt.xlabel('Caras del Dado')
plt.ylabel('Probabilidad')
plt.title('Distribución de Probabilidad (Dado de 6 Caras)')
plt.show()
```

```python
# Aleatorio
# número de lanzamientos
num_lanzamientos = 1000

# Simular los lanzamientos del dado
lanzamientos = np.random.randint(1, 7, num_lanzamientos)

# frecuencias con la que sale cada cara del dado
frecuencias = np.bincount(lanzamientos)[1:]

# calcular la probabilidad de cada cara de acuerdo a su frecuencia respecto al número de lanzamientos
probabilidades = frecuencias / num_lanzamientos

# Calcular la entropía (variables no conocidas)
entropia_dado = -np.sum(probabilidades * np.log2(probabilidades)) # H(x) = -sum(p(k).log2(p(k)))

# Graficar la frecuencia de cada cara
caras_dado = np.arange(1, 7)
plt.bar(caras_dado, frecuencias)
plt.xlabel('Caras del Dado')
plt.ylabel('Frecuencia')
plt.title(f'{num_lanzamientos} Lanzamientos de Dado de 6 Caras')
plt.show()

# mostrar la entropía del dado
print(f'Entropía (Dado de 6 Caras (100 lanzamientos)): {entropia_dado:.5f} bits')
```
